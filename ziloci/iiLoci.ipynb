```python
%matplotlib inline
from __future__ import print_function
import seaborn
import matplotlib.pyplot as plt
import numpy
seaborn.set_context({"figure.figsize": (12, 6)})
seaborn.set_style('dark')
```

```python
def plot_hist_rug(data, bins=None, xlim=None, ylim=None, title=None,
                  xlab='iiLocus Length (bp)', ylab='Frequency', xlog=True):
    """Boilerplate code for visualizations."""
    if bins is None and xlim is not None:
        bins = numpy.arange(xlim[0], xlim[1], 0.2)
    if xlog:
        plot = seaborn.distplot(numpy.log10(data), bins=bins, kde=False, rug=True)
    else:
        plot = seaborn.distplot(data, bins=bins, kde=False, rug=True)
    seaborn.plt.xlabel(xlab, fontsize=16)
    seaborn.plt.ylabel(ylab, fontsize=16)
    seaborn.despine(left=True)
    if title:
        seaborn.plt.title(title, fontsize=18)
    if xlim is not None:
        _ = plot.set(xlim=(xlim[0], xlim[1]))
    if ylim is not None:
        _ = plot.set(ylim=(ylim[0], ylim[1]))
```

```python
def parse_iilocus_lengths(infile):
    for line in infile:
        length = int(line.rstrip().split('\t')[1])
        yield length
```

```python
def iilocus_hist_orig_vs_shuf(species, xmax, ymax):
    plt.subplot(121)
    filename = '../species/%s/shuffled/%s.orig.ilens.tsv' % (species, species)
    with open(filename, 'r') as infile:
        alllengths = [x for x in parse_iilocus_lengths(infile)]
        lengths = [x for x in alllengths if x > 0]
        avg = numpy.mean(lengths)
        med = numpy.median(lengths)
        plt.title('As annotated (%d/%d > 0; avg=%.1f, med=%d)' % (len(lengths), len(alllengths), avg, med))
        plot_hist_rug(lengths, xlab='Length (bp, log base 10)',
                      xlim=(2, xmax), ylim=(0, ymax))
        plt.axvline(numpy.log10(med), color='#ff0000')
        plt.axvline(numpy.log10(avg), color='#00cc00')

    plt.subplot(122)
    filename = '../species/%s/shuffled/%s.shuffled.ilens.tsv' % (species, species)
    with open(filename, 'r') as infile:
        alllengths = [x for x in parse_iilocus_lengths(infile)]
        lengths = [x for x in alllengths if x > 0]
        avg = numpy.mean(lengths)
        med = numpy.median(lengths)
        plt.title('Shuffled (%d/%d > 0; avg=%.1f, med=%d)' % (len(lengths), len(alllengths), avg, med))
        plot_hist_rug(lengths, xlab='Length (bp, log base 10)', ylab='',
                      xlim=(2, xmax), ylim=(0, ymax))
        plt.axvline(numpy.log10(med), color='#ff0000')
        plt.axvline(numpy.log10(avg), color='#00cc00')

    plt.suptitle('%s iiLocus Lengths' % species, fontsize=16)
    plt.show()
```

```python
specdata = [('Amel', 6.0, 30), ('Bter', 6.0, 32),
            ('Hsal', 6.0, 39), ('Cflo', 6.0, 48),
            ('Pdom', 6.0, 60), ('Nvit', 6.0, 55),
            ('Atha', 6.0, 130), ('Bdis', 6.0, 70),
            ('Osat', 6.0, 70), ('Dmel', 6.0, 92),
            ('Xtro', 6.0, 25), ('Drer', 6.0, 18),
            ('Btau', 6.0, 18), ('Mmus', 6.0, 24)]
for species, xmax, ymax in specdata:
    iilocus_hist_orig_vs_shuf(species, xmax, ymax)
```
