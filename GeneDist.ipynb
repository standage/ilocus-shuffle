```python
%matplotlib inline
from __future__ import print_function
import seaborn
import matplotlib.pyplot as plt
seaborn.set_context({"figure.figsize": (12, 8)})
```

```python
def plot_hist_rug(data, bins=30, xlim=None, ylim=None, title=None,
                  xlab='Genomic position', ylab='Frequency'):
    """Boilerplate code for visualizations."""
    plot = seaborn.distplot(data, bins=bins, kde=False, rug=True)
    seaborn.plt.xlabel(xlab, fontsize=16)
    seaborn.plt.ylabel(ylab, fontsize=16)
    if title:
        seaborn.plt.title(title, fontsize=18)
    if xlim is not None:
        _ = plot.set(xlim=(xlim[0], xlim[1]))
    if ylim is not None:
        _ = plot.set(ylim=(ylim[0], ylim[1]))
```

```python
with open('species/Pdom/shuffled/Pdom.orig.iloci.gff3', 'r') as instream:
    starts = list()
    for line in instream:
        if '\tgene\t' in line:
            start = int(line.split('\t')[3])
            starts.append(start)
    plot_hist_rug(starts,
                  title='Gene distribution (before shuffling)')
```

```python
with open('species/Pdom/shuffled/Pdom.shuffled.iloci.gff3', 'r') as instream:
    starts = list()
    for line in instream:
        if '\tgene\t' in line:
            start = int(line.split('\t')[3])
            starts.append(start)
    plot_hist_rug(starts,
                  title='Gene distribution (after shuffling)')
```
